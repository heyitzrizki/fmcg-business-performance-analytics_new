# 9. Product Portfolio Segmentation

```python
# Create product portfolio base
product_portfolio = product_kpi.copy()

revenue_threshold = product_portfolio["revenue_share"].median()
margin_threshold = product_portfolio["profit_margin"].median()

print("Revenue share threshold:", revenue_threshold)
print("Profit margin threshold:", margin_threshold)

def assign_portfolio_segment(row):
    high_revenue = row["revenue_share"] >= revenue_threshold
    high_margin = row["profit_margin"] >= margin_threshold
    
    if high_revenue and high_margin:
        return "Core Growth"
    elif high_revenue and not high_margin:
        return "Margin Improvement"
    elif not high_revenue and high_margin:
        return "Niche Opportunity"
    else:
        return "Review / Rationalize"

product_portfolio["portfolio_segment"] = product_portfolio.apply(assign_portfolio_segment, axis=1)

product_portfolio["strategic_action"] = product_portfolio["portfolio_segment"].map({
    "Core Growth": "Protect availability, sustain investment, and use as core growth driver.",
    "Margin Improvement": "Review pricing, discounting, COGS, and promotion efficiency.",
    "Niche Opportunity": "Explore targeted growth opportunities with selective investment.",
    "Review / Rationalize": "Review SKU role, reduce inefficiency, or consider rationalization.",
})

product_portfolio.head()
```

```python
# Create category portfolio segmentation
category_portfolio = category_kpi.copy()

category_revenue_threshold = category_portfolio["revenue_share"].median()
category_margin_threshold = category_portfolio["profit_margin"].median()

def assign_category_segment(row):
    high_revenue = row["revenue_share"] >= category_revenue_threshold
    high_margin = row["profit_margin"] >= category_margin_threshold
    
    if high_revenue and high_margin:
        return "Core Growth"
    elif high_revenue and not high_margin:
        return "Margin Improvement"
    elif not high_revenue and high_margin:
        return "Niche Opportunity"
    else:
        return "Review / Rationalize"

category_portfolio["portfolio_segment"] = category_portfolio.apply(assign_category_segment, axis=1)

category_portfolio["strategic_action"] = category_portfolio["portfolio_segment"].map({
    "Core Growth": "Scale and defend category leadership.",
    "Margin Improvement": "Improve margin through pricing, cost, promotion, or mix optimization.",
    "Niche Opportunity": "Test selective growth and targeted marketing.",
    "Review / Rationalize": "Review strategic fit and resource allocation.",
})

category_portfolio = category_portfolio.sort_values(
    ["portfolio_segment", "net_revenue"],
    ascending=[True, False]
).reset_index(drop=True)

category_portfolio
```

```python
# Portfolio segment summary
portfolio_segment_summary = (
    product_portfolio.groupby("portfolio_segment", as_index=False)
    .agg(
        sku_count=("SKU", "nunique"),
        net_revenue=("net_revenue", "sum"),
        profit=("profit", "sum"),
        units_sold=("units_sold", "sum"),
        avg_profit_margin=("profit_margin", "mean"),
        avg_marketing_efficiency=("marketing_efficiency", "mean"),
    )
)

portfolio_segment_summary["revenue_share"] = (
    portfolio_segment_summary["net_revenue"] / portfolio_segment_summary["net_revenue"].sum()
)

portfolio_segment_summary["profit_share"] = (
    portfolio_segment_summary["profit"] / portfolio_segment_summary["profit"].sum()
)

portfolio_segment_summary = portfolio_segment_summary.sort_values(
    "net_revenue",
    ascending=False
).reset_index(drop=True)

portfolio_segment_summary
```

# 10. Profitability Review Artifacts

```python
# Negative profit transactions review
negative_profit_transactions = fmcg_cleaned[fmcg_cleaned["Profit_USD"] < 0].copy()

negative_profit_summary = (
    negative_profit_transactions.groupby(
        ["Product_Category", "Sales_Channel", "Promotion_Type"],
        as_index=False
    )
    .agg(
        loss_transaction_count=("Order_ID", "nunique"),
        net_revenue=("Net_Revenue_USD", "sum"),
        total_loss=("Profit_USD", "sum"),
        avg_profit_margin=("Profit_Margin", "mean"),
        avg_discount_pct=("Discount_Pct", "mean"),
        marketing_spend=("Marketing_Spend_USD", "sum"),
        logistics_cost=("Logistics_Cost_USD", "sum"),
    )
)

negative_profit_summary = negative_profit_summary.sort_values(
    "total_loss",
    ascending=True
).reset_index(drop=True)

negative_profit_summary.head(20)
```
